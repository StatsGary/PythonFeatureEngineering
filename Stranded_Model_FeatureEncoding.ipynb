```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://hutsons-hacks.info/wp-content/uploads/2021/04/LongLOSData.csv")
```

```python
df
```

```python
df.dropna()
```

```python
df.dtypes
```

```python
# Select only the categorical variables
cat_df = df.select_dtypes(include=['object']).copy()
```

```python
cat_df
```

```python
# Label encoding
cat_df = cat_df.drop(['stranded.label', 'admit_date'], axis=1)
```

```python
cat_df
```

```python
cat_df['frailty_index'] = cat_df['frailty_index'].astype('category')
```

```python
cat_df.dtypes
```

```python
# Create the category encoding
cat_df['frail_cat_code'] = cat_df['frailty_index'].cat.codes
```

```python
cat_df
```

```python
# One hot encoding - dummy variable encoding
dummies = pd.get_dummies(cat_df, columns=['frailty_index'])
dummies
```

```python
combined = pd.concat([df, dummies], axis=1)
combined
```

```python
final_df = combined.drop(['admit_date', 'frailty_index', 'frail_cat_code'], axis=1)
```

```python
final_df
```

```python
# Create custom encodings
cat_df['frailty_index'].value_counts()
```

```python
cat_df['mobility_issues'] = np.where(cat_df['frailty_index'].str.contains("Mobility problems"), 1,0)
```
